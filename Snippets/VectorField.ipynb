```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
```

```python
xo, yo = np.meshgrid(x,y)
```

```python
# This is the vector field generated by the differential equations system
# dx/dt = y, dy/dt = sin(x) 

q = plt.quiver(xo,yo, yo, -np.sin(xo))
```
