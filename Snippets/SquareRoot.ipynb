```python
def raiz(a,tol):
    x = a
    x2 = (x + a / x) / 2
    while abs(x-x2) >= tol:
      x=x2
      x2 = (x + a / x) / 2
    return x2
```

```python
raiz(2,1)
```

```python
raiz(2,.2)
```

```python
raiz(2,.1)
```

```python
raiz(2,.01)
```

```python
raiz(2,0.001)
```
